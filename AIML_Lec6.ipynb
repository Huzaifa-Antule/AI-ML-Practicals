```python
# For age we make use of KDE . A Kerne l density Estimate plot is a method for Visualizing the distributions of Observations in a dataset analogous to Histogram .
#  KDE Represents data using a conatianers probability density curve in one more Dimension.
```

```python
sns.kdeplot(x=DF['Age'])
```

```python
#  Box and Whiskker plots. also known as Box Plots are used to show the distribution of data points accross as selected measures ,
# They display ranges within variables measured , this include the outliers the median , mode and where the majority of the data point lie in the box
```

```python
sns.boxplot(x=DF['Age'])
```

```python
# # A Histogram is a classical Visualization tool that represents the distribution on one ore more variables by counting the number of observation tht fall within descrete bins 

# KDe this tool :if True , computes a kernel density to estimate to smooth the distribution and show on the plot as (one or more) Lines .
# It is onlu relevant with unvariant data 
```

```python
sns.histplot(x=DF['Age'],kde=True,stat='density',linewidth = 0)
```

```python
# Age of the customers Follows right skeewed normal distribution.
# A right  Skewed distribution als ocalled as positive skew distr5ibution is when the charts tail is longer on its right and its peak veers to thr left.
```

```python
# Annual Income 
sns.histplot(x=DF['Annual Income (k$)'],kde=True,stat='density',linewidth=0)
```

```python
# Annual Income of the customers also follows right skwed normal distribtion
```

```python
sns.boxplot(x=DF['Annual Income (k$)'])
```

```python
# Spending Score
sns.histplot(x=DF['Spending Score (1-100)'],kde=True,stat='density',linewidth=0)
```

```python
# Spending scre on he customers roughly follows normal distribution
```

```python
sns.boxplot(x=DF['Spending Score (1-100)'])
```

```python
# bivariate Analysyis refers to analysyis of two variables to determine reltionshp between them
# We use paiplot to chk reltionshp between difff featues
```

```python
sns.pairplot(DF[['Gender','Age','Annual Income (k$)','Spending Score (1-100)']])
```

```python
# Observations :
#  most of the customers are in the age of 20-40
# Spending sc is high for customers in the group of 20-40
# Spending sc is high for customers with very low and very high income
```

```python
# Correlation btwn Variables
DF.corr()
```

```python
sns.heatmap(DF[['Gender','Age','Annual Income (k$)','Spending Score (1-100)']].corr(),cmap='magma_r',annot= True,linewidth=0.5)
plt.title("Heat Map",fontsize=20)
plt.show()
```

```python
# Gender Vs Spending spehere
plt.rcParams['figure.figsize'] = (16,7)
g = sns.catplot(x='Gender',y = 'Spending Score (1-100)',kind='violin',inner=None,data=DF)
sns.swarmplot(x='Gender',y = 'Spending Score (1-100)',color='k',size=3,data=DF,ax=g.ax)
plt.title("Gender vs SS", fontsize=16)
plt.xlabel("Gender")
plt.ylabel("Spending Score (1-100)")
plt.show()
```

```python
# Gender Vs Annual Income
plt.rcParams['figure.figsize'] = (16,7)
g = sns.catplot(x='Gender',y = 'Annual Income (k$)',kind='violin',inner=None,data=DF)
sns.swarmplot(x='Gender',y = 'Annual Income (k$)',color='k',size=3,data=DF,ax=g.ax)
plt.title("Gender vs Annual Income (k$)", fontsize=16)
plt.xlabel("Gender")
plt.ylabel("Annual Income (k$)")
plt.show()
```

```python
# Gender Vs Age
plt.rcParams['figure.figsize'] = (16,7)
g = sns.catplot(x='Gender',y = 'Age',kind='violin',inner=None,data=DF)
sns.swarmplot(x='Gender',y = 'Age',color='k',size=3,data=DF,ax=g.ax)
plt.title("Gender vs Age", fontsize=16)
plt.xlabel("Gender")
plt.ylabel("Age")
plt.show()
```
