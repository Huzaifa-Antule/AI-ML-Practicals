<h1> K means Clustoring Based  on Annual Income </h1>

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
DF = pd.read_csv(r"C:\Users\Admin\Downloads\Mall_Customers.csv")
```

```python
DF
```

```python
X = DF[['Annual Income (k$)','Spending Score (1-100)']]
```

```python
X
```

```python
# Visualization 
plt.title("Unclustered Data ")
plt.xlabel("Annual Income ",fontsize=20)
plt.ylabel("Spending Score ",fontsize=20)
plt.grid()
plt.scatter(X['Annual Income (k$)'],X['Spending Score (1-100)'],marker='^')
plt.show()
```

```python
from sklearn.cluster import KMeans
```

```python
# elbow Method to find optimal Number of Clustors 

sse = []
for K in range(1,41):
    KM = KMeans(n_clusters=K,random_state=0,n_init='auto')
    labels = KM.fit_predict(X)
    sse.append(KM.inertia_)
```

```python
plt.figure(figsize=(16,9))
plt.title("Elbow")
plt.xlabel("Value of K ")
plt.ylabel("SSE ")
plt.grid()
plt.xticks(range(1,41))
plt.plot(range(1,41),sse,marker='o',color='r')
```

<h1> Sillhoutte Method </h1>

```python
from sklearn.metrics import silhouette_score
```

```python
silh = []

for K in range(2,16):
    KM = KMeans(n_clusters=K,random_state=0,n_init='auto')
    labels = KM.fit_predict(X)
    score = silhouette_score(X,labels)
    silh.append(score)
```

```python
# plot  the silhoutte score

plt.title("silhouette Analysyis")
plt.xlabel("Value of K ")
plt.ylabel("silhouette Score ")
plt.xticks(range(2,16))
plt.bar(range(2,16),silh,color='r')
```

```python
#  Create the Objetc
KM2 = KMeans(n_clusters=5,random_state=0,n_init='auto')
```

```python
# Train the Algorithm
labels = KM2.fit_predict(X)
```

```python
labels
```

```python
# cluster Labels 
KM2.labels_
```

```python
KM2.inertia_
```

```python
# Centriods 
KM2.cluster_centers_
```

<h1> Association Rule </h1>

```python
from mlxtend.frequent_patterns import apriori,association_rules
```

```python
DF2 = pd.read_csv(r"C:\Users\Admin\Downloads\GroceryStoreDataSet.xls",names=['products'])
DF2.head()
```

```python
DF2.shape
```

```python
# split the data and create a list 
data = list(DF2['products'].apply(lambda x: x.split(",")))
```

```python
data
```

```python
DF2['products'].str.split(',')
```

```python
# Appriori's Algorithm Transforms True/False  or 1/0 Using TransactionEncoder we convert the list to a One hot encoded boolean list.
# Products that customers bought or fif not buy during shopping will be converted to 1 to 10
```

```python
from mlxtend.preprocessing import TransactionEncoder
```

```python
# Create Ovjhect 
a = TransactionEncoder()
# apply operation to tranform
a_data = a.fit(data).transform(data)
```

```python
a_data
```

```python
a.columns_
```

```python
New_DF = pd.DataFrame(a_data,columns=a.columns_)
```

```python
New_DF
```

```python
# Converting Data Into 0-1
New_DF = New_DF.replace(False,0)
New_DF = New_DF.replace(True,1)
```

```python
New_DF
```

```python
# Applying approiri abd resulting 
# Slecting Minimum support to 20 % and print all Values 
```

```python
New_DF2= apriori(New_DF.astype(bool),min_support=0.2,use_colnames=True,verbose=1)
```

```python
New_DF2
```

```python
New_DF2.sort_values(by='support',ascending=False)
```

```python
# Putting confidence to 60%, ie when product X is purchased , we can sau that the purchase of Y is 60% or more 
```

```python
DF_AR = association_rules(New_DF2,metric='confidence',min_threshold=0.6)
```

```python
DF_AR
```
